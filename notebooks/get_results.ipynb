```python
import numpy as np
rows = []

tp = 0
fp = 0
fn = 0
tn = 0

for i, row in data.iterrows():
    is_dupicate = False
    
    if row['uuid'] in set(d.keys()):
        duplicate_for = d[row['uuid']]
        is_dupicate = True
        
    if is_dupicate == True and row['is_duplicate'] == True:
        if row['duplicate_for'] == duplicate_for:
            tp += 1
        else:
            fp += 1
    elif is_dupicate == False and row['is_duplicate'] == False:
        tn += 1
    elif is_dupicate == True and row['is_duplicate'] == False:
        fp += 1
        print(row['uuid'])
    elif is_dupicate == False and row['is_duplicate'] == True:
        fn += 1
            
```
